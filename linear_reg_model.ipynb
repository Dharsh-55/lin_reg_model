```python
import pandas as pd
epl = pd.read_excel('C:\\Users\\Admin\\Desktop\\EF 17-18 model\\epl_pyexf.xlsx')
epl['GD'] = epl['GF'] - epl['GA']
epl
```

```python
#dependent variable
y = epl['wp']
y
```

```python
#converting categorical data to numeric data
epl['div'] = epl['Div'].astype('category').cat.codes
epl['team_code'] = epl['team'].astype('category').cat.codes
epl
```

```python
#independent variable
#x = epl.drop(['wp','team','Div'], axis = 1)
x = epl[['G','W','GD', 'pyex']]
#x['GD'] = epl['GF'] - epl['GA']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
#training
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
#to predict
pred_tr = reg.predict(x_train)
pred_ts = reg.predict(x_test)
pred_tr
```

```python
pred_ts
```

```python
#model performance
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

regms_tr = mean_squared_error(y_train, pred_tr)
regr2_tr = r2_score(y_train, pred_tr)
rmse_tr = np.sqrt(mean_squared_error(y_train, pred_tr))

regms_ts = mean_squared_error(y_test, pred_ts)
regr2_ts = r2_score(y_test, pred_ts)
rmse_ts = np.sqrt(mean_squared_error(y_test, pred_ts))
```

```python
print('ACCURACY_TRAIN :')
print('mse ',regms_tr)
print('rmse', rmse_tr)
print('r2 ',regr2_tr)
print('*'*40)
print('ACCURACY_TEST :')
print('mse ',regms_ts)
print('rmse ', rmse_ts)
print('r2 ',regr2_ts)
```

```python
#train data plot
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(3,3))

plt.scatter(x = y_train, y = pred_tr, alpha = 0.3)

z = np.polyfit(y_train, pred_tr,1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), 'r', linewidth = '.5')
plt.xlabel('indp factors')
plt.ylabel('win percentage')
plt.title('train data')
plt.plot()
```

```python
#test data plot
plt.figure(figsize=(3,3))

plt.scatter(x = y_test, y = pred_ts, alpha = 0.3)

z = np.polyfit(y_test, pred_ts,1)
p = np.poly1d(z)

plt.plot(y_test, p(y_test), 'r', linewidth = '.5')
plt.xlabel('indp factors')
plt.ylabel('win percentage')
plt.title('test data')
plt.plot()
```
